## Corrections automatiques
Dans un premier temps, nous remplaçons les "id" dans la colonne chef_vocation par la vocation dans la ligne en dessus

```python
df_jobs = pd.read_csv(COMPLETE_RAW_DATA)

df_jobs = df_jobs.fillna("")

# Replace "id" by value in row above in chef_vocation
index_id = df_jobs[df_jobs["chef_vocation"].apply(lambda s: s == "id")].index

print(f"There were {len(index_id)} 'id' in 'chef_vocation'")
for idx in index_id:
    df_jobs.loc[idx, "chef_vocation"] = df_jobs.loc[idx-1,"chef_vocation"]
    
df_jobs
```

```python
df_jobs = df_jobs.fillna("").replace('·', "") # Remove default value

# Create list form "|"-separated string of job
df_jobs["chef_vocation"] = df_jobs["chef_vocation"].str.split("|")
df_jobs["enfants_profession"] = df_jobs["enfants_profession"].str.split("|")
df_jobs["pensionnaires_condition"] = df_jobs["pensionnaires_condition"].str.split("|")
df_jobs = df_jobs.fillna("").replace('·', "") # Remove default value that emerge from split
```

```python
# Merge all profession column into profession
df_jobs["profession"] = df_jobs["chef_vocation"] + df_jobs["enfants_profession"] + df_jobs["pensionnaires_condition"]
# Remove empty profession from list
df_jobs["profession"] = df_jobs["profession"].apply(lambda l: [prof for prof in l if len(prof) != 0])
```

```python
from ast import literal_eval

df_jobs = pd.read_csv(ORGANIZED_RAW_DATA)
df_jobs[["chef_vocation", "enfants_profession", "pensionnaires_condition", "profession"]] = df_jobs[["chef_vocation", "enfants_profession", "pensionnaires_condition", "profession"]].applymap(literal_eval)
df_jobs
```

Nous devons maintenant séléctionner les données à corriger manuellement.

```python
df_job_year = df_jobs[["profession", "annee"]].explode("profession").copy()
df_job_year = df_job_year[~df_job_year["profession"].isin({"·", ""})]
df_job_year = df_job_year.dropna()
df_job_year
```

Nous avons choisi de garder les top 100 métiers les plus fréquents par année.

```python
top_per_year = 100
df_job_by_year = df_job_year.groupby("annee")["profession"].apply(list)
df_count_top_by_year = df_job_by_year.apply(lambda l: Counter(l).most_common(top_per_year))

df_top_job_by_year = df_count_top_by_year.apply(lambda l: [prof for prof, count in l])
print("Métiers les plus fréquents par année:")
df_top_job_by_year
```

```python
total_occ = len(df_job_year)
print(f"Après les corrections automatiques nous avons {total_occ} occurrences de métier.")

top_terms = set(df_top_job_by_year.explode())
print(f"Il y a {len(top_terms)} termes distincts dans le top {top_per_year} termes les plus fréquents par année.")

total_kept_occ = sum(df_count_top_by_year.apply(lambda l: sum([count for prof, count in l])))
print(f"Après avoir conservé uniquement les occurrences de ces {len(top_terms)} termes distincts, nous avons {total_kept_occ} occurrences de travail.")
print(f"Cela représente {total_kept_occ / total_occ * 100}% des occurrences totales.")
```

Nous utilisons la liste des métiers corrigés de 1855 fournie par les assistants pour les enlever de la listes des métiers à corriger.

```python
df_jobs_1855 = pd.read_excel(CORRECT_TERMS_1855, index_col=0)
df_jobs_1855["prof_norm"] = df_jobs_1855["prof_norm"].apply(lambda x: x.replace("?", ""))
jobs_1855 = set(df_jobs_1855["prof_norm"])
print(f"There is {len(jobs_1855)} corrected jobs in 1885")
jobs_1855
```

```python
already_correct_terms_in_top_100 = top_terms.intersection(jobs_1855)
pd.DataFrame(already_correct_terms_in_top_100).to_csv(ALREADY_CORRECT_TERMS_TOP_100, index=False)
```

```python
terms_to_correct = top_terms - jobs_1855

print(f"Après avoir supprimé les métiers corrects, il reste {len(terms_to_correct)} termes à vérifier/corriger")

occ_terms = df_job_year[df_job_year["profession"].isin(terms_to_correct)].groupby("profession").count()
occ_terms
```
