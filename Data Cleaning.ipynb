```python
df_census_reduced = pd.read_csv(PROCESSED_DATA_DIR+'census_reduced.csv')
df_census_reduced = df_census_reduced.fillna("")

# Replace "id" by value in row above in chef_vocation
index_id = df_census_reduced[df_census_reduced["chef_vocation"].apply(lambda s: s == "id")].index
for idx in index_id:
    df_census_reduced.loc[idx, "chef_vocation"] = df_census_reduced.loc[idx-1,"chef_vocation"]
    
df_census_reduced
```

## Remove default value and separate different individual 

```python
def fillId(l):
    '''
    Return a list with "id" replaced by the previous term
    '''
    return_list = []
    prev = "id"
    for term in l:
        if term != "id":
            prev = term
        return_list.append(prev)
    return return_list
```

```python
df_census_reduced = df_census_reduced.fillna("").replace('·', "") # Remove default value

# Create list form "|"-separated string of job
df_census_reduced["chef_vocation"] = df_census_reduced["chef_vocation"].str.split("|")
df_census_reduced["enfants_profession"] = df_census_reduced["enfants_profession"].str.split("|")
df_census_reduced["pensionnaires_condition"] = df_census_reduced["pensionnaires_condition"].str.split("|")
df_census_reduced = df_census_reduced.fillna("").replace('·', "") # Remove default value that emerge from split

# Fill id (=idem) field by previous job
df_census_reduced["chef_vocation"] = df_census_reduced["chef_vocation"].apply(lambda l: fillId(l))
df_census_reduced["enfants_profession"] = df_census_reduced["enfants_profession"].apply(lambda l: fillId(l))
df_census_reduced["pensionnaires_condition"] = df_census_reduced["pensionnaires_condition"].apply(lambda l: fillId(l))
```

```python
# Merge all profession column into profession
df_census_reduced["profession"] = df_census_reduced["chef_vocation"] + df_census_reduced["enfants_profession"] + df_census_reduced["pensionnaires_condition"]
# Remove empty profession from list
df_census_reduced["profession"] = df_census_reduced["profession"].apply(lambda l: [prof for prof in l if len(prof) != 0])
```

## Reshape datastructure 

```python
df_job_year = df_census_reduced[["profession", "annee"]].explode("profession").copy()
df_job_year = df_job_year[~df_job_year["profession"].isin({"·", ""})]
df_job_year = df_job_year.dropna()
df_job_year
```

```python
df_job_by_year = df_job_year.groupby("annee")["profession"].apply(list)
print("Jobs listed by year:")
df_job_by_year
```

```python
top_per_year = 100
df_count_top_by_year = df_job_by_year.apply(lambda l: Counter(l).most_common(top_per_year))

df_top_job_by_year = df_count_top_by_year.apply(lambda l: [prof for prof, count in l])
print("Most frequent jobs by year:")
df_top_job_by_year
```

```python
total_occ = len(df_job_year)
print(f"After automatic correction we keep {total_occ} job occurrences.")

top_terms = set(df_top_job_by_year.explode())
print(f"There is {len(top_terms)} distinct terms in the {top_per_year} most frequent terms by year.")

total_kept_occ = sum(df_count_top_by_year.apply(lambda l: sum([count for prof, count in l])))
print(f"After keeping only occurrences of this {len(top_terms)} distinct terms, we have {total_kept_occ} job occurrences.")
print(f"This represent {total_kept_occ / total_occ * 100}% of the occurrences.")

```

```python
df_jobs_1855 = pd.read_excel(DATA_DIR+"metiers_1855.xlsx", index_col=0)
df_jobs_1855["prof_norm"] = df_jobs_1855["prof_norm"].apply(lambda x: x.replace("?", ""))
jobs_1855 = set(df_jobs_1855["prof_norm"])
print(f"There is {len(jobs_1855)} corrected jobs in 1885")
jobs_1855
```

```python
terms_to_correct = top_terms - jobs_1855

print(f"After removing correct jobs, there is {len(terms_to_correct)} terms to check/correct")

with open('tocorrect.csv', 'w') as f:
    writer = csv.writer(f)
    for val in sorted(list(terms_to_correct)):
        writer.writerow([val])

print("Below are the terms, with their number of occurrence:")
occ_terms = df_job_year[df_job_year["profession"].isin(terms_to_correct)].groupby("profession").count()
occ_terms
```
